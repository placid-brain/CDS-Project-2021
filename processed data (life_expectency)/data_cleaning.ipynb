```python
import pandas as pd

temp_csv = "HadCRUT Temperature Anomaly.csv"
rain_csv = "rainfall.csv"
hle_csv = "Healthy Life Expectancy.csv"

temperature = pd.read_csv(temp_csv)
rainfall = pd.read_csv(rain_csv)
expectancy = pd.read_csv(hle_csv)
```

```python
print("temperature: ")
print(temperature.head(10))
print(temperature.columns)
print(temperature.size)
print()
print("rainfall: ")
print(rainfall.head(10))
print(rainfall.columns)
print(rainfall.size)
print()
print("expectancy: ")
print(expectancy.head(10))
print(expectancy.columns)
print(expectancy.size)
print()
```

```python
params = ["Country","Year","Surface temperature anomaly","Surface temperature anomaly weighted by population","Surface temperature anomaly weighted by area","Rainfall","Healthy Life Expectancy","Life Expectancy","Years Lived With Disability"]
integrated_data = pd.DataFrame(columns = params)
for _, row in rainfall.iterrows():
  # rainfall
  country = row["Area"]
  year = row["Year"]
  rainfall_value = row["Value"]

  # expectancy
  hle = expectancy.loc[(expectancy['Entity'] == country) & (expectancy["Year"] == year)]
  if hle.size == 0:
    continue
  else:
    healthy_expectancy = hle["Healthy Life Expectancy (IHME)"].values[0]
    life_expectancy = hle["Life Expectancy (IHME)"].values[0]
    disability = hle["Years Lived With Disability (IHME)"].values[0]

    # temperature
    temp = temperature.loc[(temperature['Entity'] == country) & (temperature["Year"] == year)]
    if temp.size == 0:
      continue
    else:
      surface_temp = temp["Surface temperature anomaly"].values[0]
      surface_temp_population = temp["Surface temperature anomaly weighted by population"].values[0]
      surface_temp_area = temp["Surface temperature anomaly weighted by area"].values[0]

  df = pd.DataFrame([[country,year,surface_temp,surface_temp_population,surface_temp_area,rainfall_value,healthy_expectancy,life_expectancy,disability]],columns=params)
  integrated_data = integrated_data.append(df,ignore_index=True)
```
