```python
import pandas as pd

data = pd.read_csv("data.csv")
```

```python
data.dtypes
```

```python
country = data["Country"]
year = data["Year"]
surface_temp = data["Surface temperature anomaly"]
surface_temp_population = data["Surface temperature anomaly weighted by population"]
surface_temp_area = data["Surface temperature anomaly weighted by area"]
rainfall = data["Rainfall"]
healthy_life_expectancy = data["Healthy Life Expectancy"]
life_expectancy = data["Life Expectancy"]
years_disability = data["Years Lived With Disability"]
```

```python
# year vs temperature

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import math

countries = data["Country"].unique()
# print(countries.size)

col = 2
row = int(math.ceil(countries.size/col))

plt.style.use('seaborn-pastel')
fig, axes = plt.subplots(row,col, figsize=(20, 36))

r = 0
for i,c in enumerate(countries):
  d = data.loc[data['Country'] == c]
  sns.lineplot(ax=axes[r, i%col], data = d,x="Year",y="Surface temperature anomaly").set_title(c)
  if i%col == col-1:
    r += 1
```
